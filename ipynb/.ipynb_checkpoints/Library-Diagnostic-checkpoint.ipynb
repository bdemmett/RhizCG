```python
%%R
library(phyloseq)
library(ggplot2)
library(foreach)
library(doParallel)
library(gridExtra)
library(dplyr)
library(tidyr)
library(vegan)
```

```python
%%R
#using physeq file with sparsity of greater than 3 in 3 samples
physeqFile3in3 = '/home/bryan/RhizCG/data/MiSeq_amplicon/MergedRuns/physeq/Full-Sparsity3in3'
physeqFile_rel = '/home/bryan/RhizCG/data/MiSeq_amplicon/MergedRuns/physeq/Full-Sparsity3in3_relabund'
physeqFile_P = '/home/bryan/RhizCG/data/MiSeq_amplicon/MergedRuns/physeq/Positives'
physeq.r = '/home/bryan/RhizCG/data/MiSeq_amplicon/MergedRuns/physeq/Full-Sparsity3in3_r'
```

## Loading Physeq Objects

```python
%%R 

physeq.Full = readRDS(physeqFile3in3)
physeq.rel = readRDS(physeqFile_rel)
physeq.P =  readRDS(physeqFile_P)
physeq.rare = readRDS(physeq.r)
```

```python
%%R
physeq.P = subset_samples(physeq.P, !(PlotID %in% c("4.16_Nneg", "4.16_Npos"))) 
physeq.P = subset_samples(physeq.P, !is.na(Plant))
physeq.P
```

# Subset just bare soils to analyze change with time

```python
%%R
# subset samples to bare
bare = subset_samples(physeq.rel, Plant == "BARE")
sample_data(bare)$DAP = factor(sample_data(bare)$DAP)

# create unifrac distance matrix
registerDoParallel(cores=28)
bare.unif.dist = distance(bare, 
                      method = "unifrac", 
                      weighted = FALSE,
                      fast = TRUE, 
                      parallel = TRUE, 
                      normalized = FALSE)

#ordination of weighted unifrac
u.nmds = ordinate(bare, 
                method = "NMDS", 
               distance = bare.unif.dist)
p = plot_ordination(bare, u.nmds, justDF = TRUE)

# create weighted unifrac distance matrix
registerDoParallel(cores=28)
bare.wunif.dist = distance(bare, 
                      method = "unifrac", 
                      weighted = TRUE,
                      fast = TRUE, 
                      parallel = TRUE, 
                      normalized = FALSE)

#ordination of weighted unifrac
w.nmds = ordinate(bare, 
                method = "NMDS", 
               distance = bare.wunif.dist)

p_w = plot_ordination(bare, w.nmds, justDF = TRUE)
```

```python
%%R -w 1000 -h 500


## generating plots
p1 = ggplot(p, aes(x = NMDS1, y = NMDS2)) +
        geom_point(aes(fill=Library), size = 6, pch=21) +
        ggtitle("Unweighted UniFrac")

p2 = ggplot(p_w, aes(x = NMDS1, y = NMDS2)) +
        geom_point(aes(fill=Library), size = 6, pch=21) +
        ggtitle("Weighted UniFrac")

grid.arrange(p1, p2, ncol = 2)
```

```python
%%R
df = as(sample_data(bare), "data.frame")
d = bare.wunif.dist
bareadonis = adonis(d ~ DAP + Rep + Library, df)
bareadonis
```

## Cluster analysis to determine if Mi-Seq Run is grouping

```python
%%R
HC = hclust(bare.wunif.dist, method = "average", members = NULL)
```

```python
%%R -h 1000
library(ape)
Var = sample_data(bare)$Library   # factor variable for colors
result_phylo <- as.phylo(HC)
p = plot(result_phylo, tip.col = Var)
p
```

## Analyze only abundant community

```python
%%R
physeq.rt = filter_taxa(physeq.rel, function(x) sum(x > 0.005) > 0, TRUE)
    
Positives = taxa_names(physeq.P)
physeq.rt = prune_taxa(Positives, physeq.rt)
physeq.rt = transform_sample_counts(physeq.rt, function(x) x/sum(x))
physeq.rt
```

```python
%%R

# subset samples to bare
bare.r = subset_samples(physeq.rt, Plant == "BARE")
sample_data(bare.r)$DAP = factor(sample_data(bare.r)$DAP)

# create unifrac distance matrix
registerDoParallel(cores=28)
bare.r.unif.dist = distance(bare.r, 
                      method = "unifrac", 
                      weighted = FALSE,
                      fast = TRUE, 
                      parallel = TRUE, 
                      normalized = FALSE)

#ordination of weighted unifrac
u.nmds = ordinate(bare.r, 
                method = "NMDS", 
               distance = bare.r.unif.dist)
p = plot_ordination(bare.r, u.nmds, justDF = TRUE)

# create weighted unifrac distance matrix
registerDoParallel(cores=28)
bare.r.wunif.dist = distance(bare.r, 
                      method = "unifrac", 
                      weighted = TRUE,
                      fast = TRUE, 
                      parallel = TRUE, 
                      normalized = FALSE)

#ordination of weighted unifrac
w.nmds = ordinate(bare.r, 
                method = "NMDS", 
               distance = bare.r.wunif.dist)

p_w = plot_ordination(bare.r, w.nmds, justDF = TRUE)
```

```python
%%R -w 1000 -h 500


## generating plots
p1 = ggplot(p, aes(x = NMDS1, y = NMDS2)) +
        geom_point(aes(fill=Library), size = 6, pch=21) +
        ggtitle("Unweighted UniFrac - abundant community \n bare soil plots")

p2 = ggplot(p_w, aes(x = NMDS1, y = NMDS2)) +
        geom_point(aes(fill=Library), size = 6, pch=21) +
        ggtitle("Weighted UniFrac - abundant community \n bare soil plots")

grid.arrange(p1, p2, ncol = 2)
```
